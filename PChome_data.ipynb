```python
import urllib.request as req, json, re, string, unicodedata
from urllib.parse import quote
```

```python
query = "抽取衛生紙"
pages = 5
prod_ids = []
prod_specs = []

for page in range(1, pages):
    url = "https://ecshweb.pchome.com.tw/search/v3.3/all/results?q={}&page={}&sort=sale/dc".format(query, page)
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36"
    }

    # 轉換 query
    trans = quote(url, safe=string.printable)
    request = req.Request(trans, headers=headers)
    with req.urlopen(request) as res:
        data = res.read().decode("utf-8")
    data = json.loads(data)
    prods = data["prods"]
    
    # 收錄所有產品 Id 進 prod_ids list
    for prod in prods:
        prod_id = prod["Id"]
        prod_ids.append(prod_id)
```

```python
# no = 1

for prod_id in prod_ids:
#     print("*** Prod", no, "***")
    ecapi = "https://ecapi-cdn.pchome.com.tw/cdn/ecshop/prodapi/v2/prod/{}/desc&fields=Id,Stmt,Equip,Remark,Liability,Kword,Slogan,Author,Transman,Pubunit,Pubdate,Approve&_callback=jsonp_desc?_callback=jsonp_desc".format(prod_id)
#     print("1. Api:", ecapi)
    api_request = req.Request(ecapi, headers=headers)
    with req.urlopen(api_request) as api_res:
        api_data = api_res.read().decode("utf-8")
    
    # api_data 清洗
    api_data = api_data.replace("try{jsonp_desc(", "")
    api_data = api_data.replace(");}catch(e){if(window.console){console.log(e);}}", "")
    api_data = json.loads(api_data)
    
#     print("2. Id:", api_data[prod_id]["Id"])
    spec = api_data[prod_id]["Stmt"]
    spec = re.sub(r"\s+", " ", spec)
    
    print(spec)
    print("-"*100)
```
